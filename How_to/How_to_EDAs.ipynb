```
#importando bibliotecas para tratar os dados
import numpy as np
import pandas as pd
```

```
#Chamando arquivo de dados com biblioteca "pandas"
df = pd.read_csv("/content/bd_pre_pos.csv", sep=";", index_col=0) #chama csv, separa por ";" as colunas e transforma a 1 col em index
df #visualizar a tabela aqui chamada de "dataframe" = df
```

# Pandas Profile

```
#Atualizando pacote da biblioteca de análise
!pip install pandas_profiling --upgrade
```

```
#importanto pacote "ProfileReport" da biblioteca "pandas_profiling"
from pandas_profiling import ProfileReport
```

```
#chamando pacote para fazer análise exploratório dos dados
profile = ProfileReport(df, title="Pandas Profiling Report Disf_Eret", explorative=True)
```

```
profile.to_file("Report_PandasProfile_Disf_Eret.html") #exportar como página da internet os resultados
```

```
#visualizando análise exploratório dos dados
profile 
```

#  SweetViz


```
#instalando bibliotteca
!pip install sweetviz
```

```
# importando sweetviz com apelido sv
import sweetviz as sv
```

```
#analisando os dados
report = sv.analyze(df)
```

```
#visualizando análise exploratório dos dados
report.show_notebook("Report_SweetViz_Disf_Eret") #alternativas => #, w="100%", h="FULL", scale=0.8) #(w="900%", h=300, scale=d1.0)
```

```
#exportar como página da internet os resultados
report.show_html('Report_SweetViz_Disf_Eret_vertical.html',layout='widescreen') #alternative ,layout='vertical')
```

```
#exportar como página da internet os resultados
report.show_html('Report_SweetViz_Disf_Eret_vertical.html',layout='vertical')
```
