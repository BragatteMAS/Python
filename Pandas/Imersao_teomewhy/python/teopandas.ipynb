#### If, Else, Elif

```python
idade = input("Entre com a sua idade")
idade = int(idade)
if idade <18:
    print("Beba leite ou suco preferencialmente.")
elif idade <= 65:
    print("Beba com parcimônia e se beber não dirija.")
else:
    print("Não ingerir bebida alcóolica com medicamentos.")
```

#### For & While

```python
for i in range(10):
    if i % 2 == 0:
        print(i, "é par")
    else:
        print(i, "é impar")
```

```python
while True:
    print("Esse não para!")
```

```python
i  = 1
while True:
    print(i)
    if i == 10:
        break
    i+=1 #mesmo que i=i+1
```

```python
#alternativa  
i = 1
while i <= 10:
    print(i)
    i += 1
```
